```python
import csv
import pprint
import uuid
import random
import numpy as np
import time

pp = pprint.PrettyPrinter(indent=4)
```

```python
csv_questions = []
with open('questions.csv', 'rb') as csvfile:
    question_reader = csv.reader(csvfile, delimiter=',')
    row_ct = 0
    for row in question_reader:
        if row_ct == 0:
            col_labels = row
        if row_ct > 0:
            question_dict = {"_id": str(uuid.uuid4())} # add _id for reference later
            element_ct = 0
            for element in row:
                if col_labels[element_ct] == "order":
                    question_dict[col_labels[element_ct]] = int(element)
                else:
                    question_dict[col_labels[element_ct]] = str(unicode(element, 'ascii', 'ignore'))
                element_ct += 1
            csv_questions.append(question_dict)
        row_ct += 1

csvfile.close()
csv_questions = sorted(csv_questions, key=lambda k: k['order']) # ordered by initial ordering
```

```python
num_difficulty = {"easy": 0, "medium": 1, "hard": 2}

list_groups = list(set([str(question['group']) for question in csv_questions]))
num_groups = {}
group_ct = 0
for group in list_groups:
    num_groups[group] = group_ct
    group_ct += 1

def initMatrix():
    database_matrix = np.zeros((len(csv_questions),8), dtype=object)
    for question_ct in range(len(csv_questions)):
        
        # Column 0 = Question ID
        database_matrix[question_ct][0] = csv_questions[question_ct]["_id"]
        
        # Column 1 = Author-Defined Order
        database_matrix[question_ct][1] = int(csv_questions[question_ct]["order"])
        
        # Column 2 = Author-Defined Difficulty
        database_matrix[question_ct][2] = num_difficulty[str(csv_questions[question_ct]["difficulty"])]
        
        # Column 3 = Author-Defined Content Group
        database_matrix[question_ct][3] = num_groups[str(csv_questions[question_ct]["group"])]
        
        # Column 4 = Answered Last Session
        # Yes and correctly = 2, Yes and incorrectly = 1, No answered last session = 0
        
        # Column 5 = Total Times Answered Correctly
        # Numeric
        
        # Column 6 = Total Times Answered
        # Numeric
        
        # Column 7 = Time Question Was Last Seen
        # Numeric
        
    return database_matrix

data_matrix = initMatrix()
print data_matrix
```

```python
#time total of studying, time they have today, rate of answering questions, probability of correctness, their database_matrix

#Askii User Class:    
class User:
    def __init__(self, name, time_today, total_time, rate_answer, prob_correct, data_matrix, total_questions_seen=0, last_question_seen=0, questions_answered_this_session=0, current_group_num=0, questions_seen_in_group=[]): 
    #Time today and total time are in minutes. Rate answer is in "how many minutes to answer a question"
        self.name = name
        self.time_today = time_today
        self.total_time = total_time
        self.rate_answer = rate_answer
        self.prob_correct = prob_correct
        self.data_matrix = data_matrix
        self.total_questions_seen = total_questions_seen
        self.last_question_seen = last_question_seen
        self.questions_answered_this_session = questions_answered_this_session
        self.current_group_num = current_group_num
        self.questions_seen_in_group = questions_seen_in_group
    
    def getQuestionRow(self, question_id):
        return np.where(self.data_matrix [:, 0] == question_id)[0][0]
    
    def getQuestionContent(self, question_row):
        return self.data_matrix[question_row]
    
    def answerQuestion(self, question_id, response, review=False):
        answered = {True: 2, False: 1}
        question_ct = self.getQuestionRow(question_id)
        if response == True and self.data_matrix[question_ct][2] > 0:
            self.data_matrix[question_ct][2] -= 1
        elif response == False and self.data_matrix[question_ct][2] < 3:
            self.data_matrix[question_ct][2] += 1
        self.data_matrix[question_ct][4] = answered[response]
        if response == True:
            self.data_matrix[question_ct][5] += 1
        self.data_matrix[question_ct][6] += 1
        self.data_matrix[question_ct][7] = time.time()
        self.last_question_seen = question_ct
        self.total_questions_seen += 1
        self.questions_answered_this_session += 1
        if review == True:
            self.questions_seen_in_group.append(question_ct) # add the question to the list of questions from the theme that the user has seen
            all_questions_in_group = np.where(self.data_matrix [:, 3] == self.current_group_num)[0]
            possible_questions = len(np.delete(all_questions_in_group, self.questions_seen_in_group))
            if possible_questions == 0:
                if self.current_group_num != max(set(self.data_matrix[:,3])):
                    # increment group number if possible and reset self.questions_seen_in_group to []
                    self.current_group_num += 1
                    self.questions_seen_in_group = []
                else:
                    # this is where we have run out of themed questions
                    pass
                
                
        return self.data_matrix[question_ct]
    
    def probabilityOfNewQuestion(self, prob_new):
        prob_review = 1.0 - prob_new
        weighted_choices = [(True, prob_new), (False, prob_review)]
        choices, weights = zip(*weighted_choices)
        return np.random.choice(choices, p=weights)
    
    def getAllPossibleReviewQuestions(self):
        # get all questions to review
        # get all rows with col[6] != 0
        row_cts = np.where(self.data_matrix [:, 6] != 0)[0]
        return row_cts
    
    def leitnerBoxSelection(self, possible_review_questions):
        
        possible_review_rows = self.data_matrix[possible_review_questions]
        
        easy_weight = 0.1
        medium_weight = 0.2
        hard_weight = 0.3
        very_hard_weight = 0.4
        
        easy_questions = np.where(possible_review_rows [:, 2] == 0)[0]
        medium_questions = np.where(possible_review_rows [:, 2] == 1)[0]
        hard_questions = np.where(possible_review_rows [:, 2] == 2)[0]
        very_hard_questions = np.where(possible_review_rows [:, 2] == 3)[0]
        
        weighted_choices = [(easy_questions, easy_weight), (medium_questions, medium_weight), 
                            (hard_questions, hard_weight), (very_hard_questions, very_hard_weight)]
        
        choices, weights = zip(*weighted_choices)
        
        chosen_question_catagory = []
        while len(chosen_question_catagory) == 0:
            chosen_question_catagory = np.random.choice(choices, p=weights)
        
        return random.choice(chosen_question_catagory) # returns row index of question to review
        
    
    def getNextQuestion(self):
        if self.total_questions_seen <= -1: #self.data_matrix.shape[0]:
            if self.questions_answered_this_session < 2:
                # if this is the first question in the session, you must ask a new question
                question_id = self.getQuestionContent(self.last_question_seen+1)[0] 
                print question_id
            else:
                # if this isn't the first question in the session, you can ask a review question
                type_question = self.probabilityOfNewQuestion(0.5)
                if type_question == True:
                    # get the next new question in order
                    question_id = self.getQuestionContent(self.last_question_seen+1)[0] 
                    print 'new ordered question: ' + question_id
                else:
                    # select a review question
                    possible_review_questions = self.getAllPossibleReviewQuestions()
                    question_ct = self.leitnerBoxSelection(possible_review_questions)
                    question_id = self.getQuestionContent(question_ct)[0]
                    # use the leitner box to select which question to review
                    print 'review question: ' + question_id
                
        else:
            print 'thematic questions'
            all_questions_in_group = np.where(self.data_matrix [:, 3] == self.current_group_num)[0]
            possible_questions = np.delete(all_questions_in_group, self.questions_seen_in_group)
            question_ct = random.choice(possible_questions)
            question_id = self.data_matrix[question_ct][0]
            if self.questions_answered_this_session < 2:
                print 'new thematic question: ' + question_id
            else:
                type_question = self.probabilityOfNewQuestion(0.5) 
                # this 0.5 value should change to be some function of how far along in the content you are
                if type_question == True:
                    # get the next new question in order
                    print 'new thematic question: ' + question_id
                else:
                    # select a review question
                    possible_review_questions = self.getAllPossibleReviewQuestions()
                    question_ct = self.leitnerBoxSelection(possible_review_questions)
                    question_id = self.getQuestionContent(question_ct)[0]
                    # use the leitner box to select which question to review
                    print 'review [in thematic] question: ' + question_id
        
    
testUser = User("Askd", 10, 50, 2, 0.75, data_matrix) #regenerating a user erases their answerng history
testUser.answerQuestion('caf8fef4-37d8-4e8b-ab0e-0a2ddbeb711e', True)
testUser.answerQuestion('b5186b2b-24f6-44a2-89e9-5d7dc2107785', True)
testUser.getNextQuestion()
```
